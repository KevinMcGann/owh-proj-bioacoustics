```python
import xarray as xr
import numpy as np
```

```python
x = xr.DataArray(range(0,100),coords={'depth':range(0,100)})
```

```python
x
```

```python
windows = x.rolling(depth=11,center=True,min_periods=11).construct('window')
windows[0:5]
```

```python
#get all the below data
windows_below = windows[:,6:]
windows_below
```

```python
bellow=windows_below.rolling(window=5,min_periods=3).mean()
bellow

```

```python
windows_above = windows[:,[4,3,2,1,0]]
windows_above
```

```python
above=windows_above.rolling(window=5,min_periods=3).mean()
above
```

```python
# swap it back round so that 1st position = max number of items   5 4 3 2 1
# first position in below = 1 item                                1 2 3 4 5
# note that less than 2 items can't be and we need 3 then only 3 3 works
bove=above[:,[4,3,2,1,0]]
above
```
